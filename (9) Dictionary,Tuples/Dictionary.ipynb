```python
people = {
    #? Item(1)
    'P1':{
        'name':'Ahmed',
        'age':20,
        'HairColor':'Brown',
        },
    
    #? Item(1)
    'P2':{
        'name':'Ahmed',
        'age':18,
        'HairColor':'Black',
        },
    
    #? Item(1)
    'P3':{
        'name':'Abeer',
        'age':25,
        'HairColor':'Red'
        },

}


# print(people.get('P1'))
person1 = people.get('P1')
print('[#] The First Person Details: \n')
print(f"[$] Name: {person1.get('name')}")
print(f"[$] Age: {person1.get('age')}")
print(f"[$] Hair Color: {person1.get('HairColor')}")
```

```python
# Two-Dimensional Dictionary
languages = {
   "One": {
      "name": "Html",
      "progress": "80%",
      },

   "Two": {
      "name": "Css",
      "progress": "90%"
   },

   "Three": {
      "name": "Js",
      "progress": "90%"
   },
}

# The First Languge:
   # name  --> Html
   # progress --> 80%

# The Second Languge:
   # name  --> Css
   # progress --> 90%

# The Third Languge:
   # name  --> Js
   # progress --> 90%

# print(languages.get('One'))
print('[$] The First Language: \n')
print(f'   [#] Name --> {languages.get("One").get("name")}')
print(f'   [#] Progress --> {languages.get("One").get("progress")}')

print('\n[$] The Second Language: \n')
print(f'   [#] Name --> {languages.get("Two").get("name")}')
print(f'   [#] Progress --> {languages.get("Two").get("progress")}')

print('\n[$] The Third Language: \n')
print(f'   [#] Name --> {languages.get("Three").get("name")}')
print(f'   [#] Progress --> {languages.get("Three").get("progress")}')
```
